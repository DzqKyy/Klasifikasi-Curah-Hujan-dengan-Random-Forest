### Cleaning Data

#### Mengubah tipe data object ke float dan date time

```python
# Mengubah tipe data ke float
all_df['Tn'] = pd.to_numeric(all_df['Tn'], errors='coerce')
all_df['Tx'] = pd.to_numeric(all_df['Tx'], errors='coerce')
all_df['Tavg'] = pd.to_numeric(all_df['Tavg'], errors='coerce')
all_df['RH_avg'] = pd.to_numeric(all_df['RH_avg'], errors='coerce')
all_df['RR'] = pd.to_numeric(all_df['RR'], errors='coerce')
all_df['ss'] = pd.to_numeric(all_df['ss'], errors='coerce')
all_df['ff_x'] = pd.to_numeric(all_df['ff_x'], errors='coerce')
all_df['ddd_x'] = pd.to_numeric(all_df['ddd_x'], errors='coerce')
all_df['ff_avg'] = pd.to_numeric(all_df['ff_avg'], errors='coerce')

# Mengubah tipe data ke datatime
all_df['Tanggal'] = pd.to_datetime(all_df['Tanggal'], format='%d-%m-%Y', errors='coerce')
```

```python
all_df.info()
```

#### Mengatasi missing value dan Mengubah nilai pada kolom RR yang bernilai 8888

```python
# Menambahkan kolom 'Bulan' dan 'Tahun'
all_df['Bulan'] = all_df['Tanggal'].dt.month
all_df['Tahun'] = all_df['Tanggal'].dt.year

# Mengonversi ke bentuk numerik (opsional)
all_df['Bulan'] = all_df['Bulan'].astype(int)
all_df['Tahun'] = all_df['Tahun'].astype(int)

# Mengganti nilai null dan 8888 dengan NaN
all_df['RR'] = all_df['RR'].replace([None, 8888], float('nan'))

# Menghitung rata-rata per bulan dan tahun
avg_Tn = all_df.groupby(['Tahun', 'Bulan'])['Tn'].transform(lambda x: x.fillna(x.mean()))
avg_Tx = all_df.groupby(['Tahun', 'Bulan'])['Tx'].transform(lambda x: x.fillna(x.mean()))
avg_Tavg = all_df.groupby(['Tahun', 'Bulan'])['Tavg'].transform(lambda x: x.fillna(x.mean()))
avg_RH_avg = all_df.groupby(['Tahun', 'Bulan'])['RH_avg'].transform(lambda x: x.fillna(x.mean()))
avg_RR = all_df.groupby(['Tahun', 'Bulan'])['RR'].transform(lambda x: x.fillna(x[x != 8888].mean()))
avg_ss = all_df.groupby(['Tahun', 'Bulan'])['ss'].transform(lambda x: x.fillna(x.mean()))
avg_ff_x = all_df.groupby(['Tahun', 'Bulan'])['ff_x'].transform(lambda x: x.fillna(x.mean()))
avg_ddd_x = all_df.groupby(['Tahun', 'Bulan'])['ddd_x'].transform(lambda x: x.fillna(x.mean()))
avg_ff_avg = all_df.groupby(['Tahun', 'Bulan'])['ff_avg'].transform(lambda x: x.fillna(x.mean()))

# Menghitung modus ddd_car per bulan dan tahun
mode_ddd_car = all_df.groupby(['Tahun', 'Bulan'])['ddd_car'].transform(lambda x: x.mode()[0] if not x.mode().empty else None)

# Mengisi nilai yang hilang dengan rata-rata
all_df['Tn'] = all_df['Tn'].fillna(avg_Tn)
all_df['Tx'] = all_df['Tx'].fillna(avg_Tx)
all_df['Tavg'] = all_df['Tavg'].fillna(avg_Tavg)
all_df['RH_avg'] = all_df['RH_avg'].fillna(avg_RH_avg)
all_df['RR'] = all_df['RR'].fillna(avg_RR)
all_df['ss'] = all_df['ss'].fillna(avg_ss)
all_df['ff_x'] = all_df['ff_x'].fillna(avg_ff_x)
all_df['ddd_x'] = all_df['ddd_x'].fillna(avg_ddd_x)
all_df['ff_avg'] = all_df['ff_avg'].fillna(avg_ff_avg)

# Mengisi nilai yang hilang dengan modus
all_df['ddd_car'] = all_df['ddd_car'].fillna(mode_ddd_car)
```
